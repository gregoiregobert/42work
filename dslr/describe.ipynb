```python
def my_count(serie):
    n = 0
    serie = serie.dropna()
    for _ in serie:
        n += 1
    return n
```

```python
def my_min(serie):
    min = serie[0]
    for i in serie:
        if i < min:
            min = i
    return min
```

```python
def my_max(serie):
    max = serie[0]
    for i in serie:
        if i > max:
            max = i
    return max
```

```python
def quantile(serie, percent):
    serie = serie.dropna()
    size = my_count(serie)
    quart = percent * (size - 1)
    serie = serie.sort_values().reset_index(drop=True)

    lower = int(quart)
    upper = lower + 1
    fraction = quart - lower

    if upper < size:
        # Interpolation lineaire : on va rajouter a la valeur en dessous du pourcentage la fraction manquante : (i[26%] - i[24%]) * (25% - 24%)
        return serie.iloc[lower] + (serie.iloc[upper] - serie.iloc[lower]) * fraction
    else:
        return serie.iloc[lower]
```
