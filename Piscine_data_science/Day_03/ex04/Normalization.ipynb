```python
import pandas as pd

# Emplacement
folder = "/Users/gregoiregobert/Downloads/42/knight"
folder = "/home/ggobert/Downloads"

knight_skills = pd.read_csv(folder + "/Test_knight.csv")
df = pd.read_csv(folder + "/Train_knight.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler

df1 = df.drop(columns=['knight'])

scaler = MinMaxScaler()
norm_df = scaler.fit_transform(df1)
norm_knghit_skills = scaler.fit_transform(knight_skills)

norm_df = pd.DataFrame(norm_df, columns=df1.columns)
norm_knight_skills = pd.DataFrame(norm_knghit_skills, columns=knight_skills.columns)

print("Standardize Train Knight: \n", norm_df)
print("\nStandardize Test Knight: \n",norm_knight_skills)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

norm_df['knight'] = df['knight']
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.scatterplot(norm_df, x='Sensitivity', y='Hability', hue='knight', ax=axes[0, 0])
sns.scatterplot(norm_knight_skills, x='Sensitivity', y='Hability', ax=axes[1, 0])
sns.scatterplot(norm_df, x='Reactivity', y='Survival', hue='knight', ax=axes[0, 1])
sns.scatterplot(norm_knight_skills, x='Reactivity', y='Survival', ax=axes[1, 1])
```
