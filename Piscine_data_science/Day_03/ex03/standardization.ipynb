```python
import pandas as pd

# Emplacement
folder = "/Users/gregoiregobert/Downloads/42/knight"

knight_skills = pd.read_csv(folder + "/Test_knight.csv")
df = pd.read_csv(folder + "/Train_knight.csv")
df1 = df
```

```python
from sklearn.preprocessing import StandardScaler

df1 = df1.drop(columns=['knight'])

scaler = StandardScaler()
std_df = scaler.fit_transform(df1)
std_knghit_skills = scaler.fit_transform(knight_skills)

std_df = pd.DataFrame(std_df, columns=df1.columns)
std_knight_skills = pd.DataFrame(std_knghit_skills, columns=knight_skills.columns)

print("Standardize Train Knight: \n", std_df)
print("\nStandardize Test Knight: \n",std_knight_skills)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

std_df['knight'] = df['knight']
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.scatterplot(std_df, x='Sensitivity', y='Hability', hue='knight', ax=axes[0, 0])
sns.scatterplot(std_knight_skills, x='Sensitivity', y='Hability', ax=axes[1, 0])
sns.scatterplot(std_df, x='Reactivity', y='Survival', hue='knight', ax=axes[0, 1])
sns.scatterplot(std_knight_skills, x='Reactivity', y='Survival', ax=axes[1, 1])
```
