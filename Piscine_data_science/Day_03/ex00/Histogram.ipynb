```python
import pandas as pd

# Emplacement
folder = "/Users/gregoiregobert/Downloads/42/knight"
folder = "/home/ggobert/Downloads"

knight_skills = pd.read_csv(folder + "/Test_knight.csv")
df = pd.read_csv(folder + "/Train_knight.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Déterminer le nombre de colonnes pour la grille
num_cols = 5
num_rows = len(df.columns) // num_cols

# Créer la figure et les axes
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 3.5))

# Parcourir chaque colonne du DataFrame et créer un histogramme
for i, column in enumerate(knight_skills.columns):
    row = i // num_cols
    col = i % num_cols
    sns.histplot(knight_skills[column], ax=axes[row, col])
    axes[row, col].set_title(f'{column}')
    axes[row, col].set_ylabel('')
    axes[row, col].set_xlabel('')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Déterminer le nombre de colonnes pour la grille
num_cols = 5
num_rows = len(df.columns) // num_cols

# Créer la figure et les axes
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 2.5))

# Parcourir chaque colonne du DataFrame et créer un histogramme
for i, column in enumerate(df.columns):
    if column != 'knight':
        row = i // num_cols
        col = i % num_cols
        sns.histplot(df, x=column, hue='knight', ax=axes[row, col])
        axes[row, col].set_title(f'{column}')
        axes[row, col].set_ylabel('')
        axes[row, col].set_xlabel('')

plt.tight_layout()
plt.show()
```
